```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df=pd.read_csv('train.txt',sep=';',header=None,names=['text','emotion'])
```

```
df.head()
```

```
df.isnull().sum()
```

```
unique_emotion=df['emotion'].unique()
```

```
emotion_number={}
i=0
for emo in unique_emotion:
    emotion_number[emo]=i
    i+=1
df['emotion']=df['emotion'].map(emotion_number)
```

```
df
```

```
df['text']=df['text'].apply(lambda x: x.lower())
```

```
df
```

```
import string
def remove_pun(txt):
  return txt.translate(str.maketrans('','',string.punctuation))
```

```
df['text']=df['text'].apply(remove_pun)
```

```
df["text"] = df["text"].str.replace(r"\d+", "", regex=True)
```

```
df["text"] = df["text"].str.replace(
    r"https?://\S+|www\.\S+",
    "",
    regex=True
)
```

```
df
```

```
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```
nltk.download('punkt')
nltk.download("punkt_tab")
nltk.download('stopwords')
df
```

```
stopw_ords = set(stopwords.words("english"))
```

```
df
```

```
df.loc[0]['text']
```

```
#function to remove stopwords
def remove(txt):
    words = word_tokenize(txt)
    cleaned = []

    for word in words:
        if word not in stopw_ords:
            cleaned.append(word)

    return " ".join(cleaned)
```

```
df['text']=df['text'].apply(remove)
```

```
df.loc[1]['text']
```

```
df
```

```
#bag of words
from sklearn.feature_extraction.text import CountVectorizer
doc=[
    "i love pizza",
    "i hate pizza",
    "i love burger",
    "i hate burger"
]
v=CountVectorizer()
x=v.fit_transform(doc)
print(v.get_feature_names_out())
print(x.toarray())
```

```
df
```

```
#bag of words
from sklearn.feature_extraction.text import TfidfVectorizer
doc=[
    "i love pizza",
    "i hate pizza",
    "i love burger",
    "i hate burger"
]
v=TfidfVectorizer()
x=v.fit_transform(doc)
print(v.get_feature_names_out())
print(x.toarray())
```

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df['text'],df['emotion'],test_size=0.2,random_state=42)
```

```
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
```

```
bow=CountVectorizer()
x_train_bow=bow.fit_transform(x_train)
x_test_bow=bow.transform(x_test)
```

```
tf=TfidfVectorizer()
x_train_tf=tf.fit_transform(x_train)
x_test_tf=tf.transform(x_test)
```

```
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```
nb=MultinomialNB()
nb.fit(x_train_bow,y_train)
y_pred=nb.predict(x_test_bow)
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test, y_pred))
```

```
tf_nb=MultinomialNB()
tf_nb.fit(x_train_tf,y_train)
y_pred=tf_nb.predict(x_test_tf)
print(accuracy_score(y_test,y_pred))
```

```
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train_bow,y_train)
y_pred=lr.predict(x_test_bow)
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test, y_pred))
```

```
lr.fit(x_train_tf,y_train)
y_pred=lr.predict(x_test_tf)
print(accuracy_score(y_test,y_pred))
```

```
from sklearn.svm import SVC
svm=SVC()
svm.fit(x_train_bow,y_train)
y_pred=svm.predict(x_test_bow)
print(accuracy_score(y_test,y_pred))
```

```
svm.fit(x_train_tf,y_train)
y_pred=svm.predict(x_test_tf)
print(accuracy_score(y_test,y_pred))
```

```
from sklearn.linear_model import SGDClassifier
sgd=SGDClassifier()
sgd.fit(x_train_bow,y_train)
y_pred=sgd.predict(x_test_bow)
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test, y_pred))
```

```
sgd.fit(x_train_tf,y_train)
y_pred=sgd.predict(x_test_tf)
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test, y_pred))
```

```
text = ["I am furious that they lied to me."]

# Convert text to TF-IDF
text_tf = tf.transform(text)

# Predict
prediction = sgd.predict(text_tf)

print("Predicted Emotion:", prediction[0])
```
